```python
def make_track_df(album_id, df_name):
    tracks_dict = genius.album_tracks(album_id)

    track_lis = []
    for ele in tracks_dict['tracks']:
        track_id = ele['song']['id']
        track_unicode = ele['song']['title']
        track_encode = track_unicode.encode("ascii","ignore")
        track_name = track_encode.decode()

        track_lis.append([track_id,track_name])
    df = pd.DataFrame(track_lis,columns=['ID','Tracks'])
    df.to_csv(df_name)
```

```python
for i in range(0,len(album_df)):
    album_name = album_df.iloc[i,1]
    album_id = album_df.iloc[i,0]
    album_df_name = re.sub('[^a-zA-Z0-9_]','_',''.join(album_name.split())) + '.csv'
    path = 'data/Tabular/'+ album_df_name
    make_track_df(album_id, path)
```

```python
coverart_path = 'data/Cover_Art'
isExist = os.path.exists(coverart_path)
if not isExist:
    os.makedirs(coverart_path)
else:
    pass
for i in range(0,len(album_df)):
    album_name = album_df.iloc[i,1]
    id = album_df.iloc[i,0]
    coverart_dirname = re.sub('[^a-zA-Z0-9_]','_',''.join(album_name.split()))   
    coverimgurl = genius.album_cover_arts(id)['cover_arts'][0]['image_url'] 
    total_path = coverart_path + "/{}".format(coverart_dirname) +'.jpg'
    imgobj = requests.get(coverimgurl).content
    f = open(total_path,'wb')
    f.write(imgobj)
    f.close()   
```
